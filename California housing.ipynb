```python
Q2 = np.median(df['HouseAge'])
Q1 = np.median(df[df['HouseAge'] < Q2]['HouseAge'])
Q3 = np.median(df[df['HouseAge'] > Q2]['HouseAge'])

print(f"count   {df['HouseAge'].count()}")
print(f"mean    {np.mean(df['HouseAge']):.6f}")
print(f"mean    {np.std(df['HouseAge']):.6f}")
print(f"min     {np.min(df['HouseAge'])}")
print(f"25%     {Q1}")
print(f"50%     {Q2}")
print(f"75%     {Q3}")
print(f"max     {np.max(df['HouseAge'])}\n")

IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
Range = [lower_bound, upper_bound]

print(f"IQR    {IQR}")
print(f"lower  {lower_bound}")
print(f"upper  {upper_bound}")
print(f"Range  {Range}")
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].hist(df['HouseAge'], bins=10, edgecolor="black", color='skyblue')
axs[0].set_title("House Age Distribution")
axs[0].set_xlabel("House Age")
axs[0].set_ylabel("Number of Houses")

axs[1].hist(df['HouseAge'], bins=50, edgecolor='black', color='yellow')
axs[1].set_title("House Age Distribution")
axs[1].set_xlabel("House Age")
axs[1].set_ylabel("Number of Houses")

plt.show()
```
